```python
import os
import six
import csv
TARGET_FILE1 =  'total'
TARGET_FILE2 = 'huff_code_test'
csv_file_name1 =TARGET_FILE1+".csv"
txt_file_name2 =TARGET_FILE2+".txt"
```

```python
def _filter_words(raw_words):
    filter_words = ""
    for word in range(len(raw_words)):
        if ord(raw_words[word]) >=256:
            filter_words += " "
        else:
            filter_words += raw_words[word]
            #print(raw_words[word])
    #print(filter_words)
    return filter_words  #也可filter_words =[],然后append
```

```python
def _letter_freq(filter_words):
    letter_freq = {}
    for word in filter_words:
        letter_freq.setdefault(word,0)  #也可采用dict=count(word)
        letter_freq[word] += 1
    #print(letter_freq)
    return letter_freq #dict
```

```python
with open("test.txt","r") as f:
    raw_words = f.read()
    #print(raw_words)
    filter_words = _filter_words(raw_words)
    letter_freq = _letter_freq(filter_words)
    print(letter_freq)
```

```python
#转为tuple
def dict2tuple(letter_freq):
    return sorted(letter_freq.items(),key = lambda x:x[1],reverse=True)
begin_tuple = dict2tuple(letter_freq)
print(begin_tuple)
```

```python
def get_coding_schedule(end1, end2, begin_tuple, code_schedule):
    if len(end1[0]) == 1:
        code_schedule.setdefault(end1[0], '1')
    else:
        for k in end1[0]:
            code_schedule[k] = '1' + code_schedule[k]
    if len(end2[0]) == 1:
        code_schedule.setdefault(end2[0], '0')
    else:
        for k in end2[0]:
            code_schedule[k] = '0' + code_schedule[k]
    begin_tuple.append((end2[0] + end1[0], end1[1] + end2[1]))
    return code_schedule
```

```python
begin_tuple = dict2tuple(letter_freq)
# 初始化哈夫曼编码表
code_schedule = {}
# 不断重排序，更新哈夫曼编码表及节点信息
for i in range(len(begin_tuple) - 1):
    begin_tuple = dict2tuple(dict(begin_tuple))
    code_schedule = get_coding_schedule(begin_tuple.pop(), begin_tuple.pop(), begin_tuple, code_schedule)
print(code_schedule)  #dict
```

```python
code = ''.join(list(code_schedule.values()))
for word in filter_words:
    code += code_schedule[word]
# 不足 8 位补 0，记录在 code_sup 中
code_sup = 8 - len(code) % 8
code += code_sup * '0'
print(code)
```

```python
with open(csv_file_name1,'w',encoding = "utf-8") as f1:
    csv_file = csv.writer(f1,delimiter = "|",quotechar='"',quoting = csv.QUOTE_MINIMAL)
    for word in letter_freq:
        num = letter_freq[word]
        huff_code = code_schedule[word]
        csv_file.writerow([word,num,huff_code]) 
```
